```python
import pandas as pd
# пандас

file = 'trekking2.xlsx'
# имя файла

sheets_dict = pd.read_excel('trekking2.xlsx', sheet_name=None)
# считываем все названия страниц в файле

sheetnames = []

for i in sheets_dict:
    sheetnames.append(i)
# добавляем все названия страниц в список    
# print(*sheetnames)

df_spr = pd.read_excel(file, sheet_name=sheetnames[0])
df_ras = pd.read_excel(file, sheet_name=sheetnames[1])
# читаем 2 страницы в разные ДатаФреймы


df_spr.columns = ['ProductSpr', 'Ccal', 'B', 'J', 'Y']
# переобзываем колонки справочника

df_ras_group_series = df_ras.groupby('Продукт')['Вес в граммах'].sum()
df_ras_group = df_ras_group_series.to_frame().reset_index()
df_ras_group.columns = ['Product', 'Weight']
# группируем список и переобзываем его колонки

df_merged = df_ras_group.merge(df_spr, left_on='Product', right_on='ProductSpr').fillna(0)
# присоединяем к списку данные из справочника, заполняем отсутствующие значения нулями

df_merged['AvgCcal'] = df_merged['Weight'] * df_merged['Ccal'] / 100
df_merged['AvgB'] = df_merged['Weight'] * df_merged['B'] / 100
df_merged['AvgJ'] = df_merged['Weight'] * df_merged['J'] / 100
df_merged['AvgY'] = df_merged['Weight'] * df_merged['Y'] / 100
# считаем средние величины

df_merged.loc['Total',:]= df_merged.sum(axis=0)
# подсчет средних значений по всем столбцам (как сделать только по последним 4-м - пока что не знаю)

df_merged
# результат
```

```python
import pandas as pd
import zipfile
import requests
import io


sheet = pd.Series()
url = 'https://stepik.org/media/attachments/lesson/245299/rogaikopyta.zip'
r = requests.get(url)

with r, zipfile.ZipFile(io.BytesIO(r.content)) as archive:
	for member in archive.infolist():
		slip = pd.read_excel(io.BytesIO(archive.read(member)))
		sheet.loc[slip.iloc[0, 1]] = int(slip.iloc[0, 3])

with open('salaries.txt', 'w') as salaries:
	salaries.write('\n'.join(sheet.sort_index().reset_index(). \
		apply(lambda x: f'{x.iloc[0]} {x.iloc[1]}', axis=1)))
```

```python
import xlrd, xlwt, requests, zipfile, io
res = requests.get(' https://stepik.org/media/attachments/lesson/245299/rogaikopyta.zip')
zfile = zipfile.ZipFile(io.BytesIO(res.content), 'r')
ww = xlwt.Workbook() 
ws = ww.add_sheet('zp')
l = []
for x in zfile.infolist():
    wb = xlrd.open_workbook(file_contents=zfile.read(x))
    sh = wb.sheet_by_name(wb.sheet_names()[0])
    l += [(sh.row_values(1)[1], sh.row_values(1)[3])]
for i, s in enumerate(sorted(l)):
    ws.write(i, 0, s[0])
    ws.write(i, 1, s[1])
ww.save('zp.xlsx')
```

```python
# Zipfile

import io
from zipfile import ZipFile
import pandas as pd

path = "rogaikopyta.zip"
zf = ZipFile(path)
vedomost = []

for file_name in zf.namelist():
    df = pd.read_excel(io.BytesIO(zf.read(file_name)))
    fio, salary = df.iloc[0, 1], df.iloc[0, 3]
    vedomost.append((fio, int(salary))

with open("vedomost.txt", "w") as f:
    for fio, salary in sorted(vedomost):
        f.write(f"{fio} {salary}\n")
```
